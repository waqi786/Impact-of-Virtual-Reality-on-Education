```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Set color palette
sns.set_palette(['#FF5733', '#337AB7', '#28A745'])  # Orange, Blue, Green
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:black; overflow:hidden"><b> ‚ú® Loading the Dataset ‚ú®</b></div>

```python
# Load dataset
data = pd.read_csv('/kaggle/input/impact-of-virtual-reality-on-education/Virtual_Reality_in_Education_Impact.csv')

# Display the first few rows of the dataset
data.head(10)
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:black; overflow:hidden"><b> ‚ú® Demographics and General Engagement ‚ú®</b></div>

```python
# Age Distribution of Students
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=10, kde=True, color='#FF5733')
plt.title('Age Distribution of Students', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gender Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=data, palette=['#337AB7', '#28A745'])
plt.title('Gender Distribution of Students', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.show()

# Engagement Level by Field of Study
plt.figure(figsize=(12, 6))
sns.boxplot(x='Field_of_Study', y='Engagement_Level', data=data, palette='pastel')
plt.title('Engagement Level by Field of Study', fontsize=16)
plt.xlabel('Field of Study', fontsize=14)
plt.ylabel('Engagement Level', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:black; overflow:hidden"><b> ‚ú® Learning Outcomes and Effectiveness ‚ú®</b></div>

```python
# Improvement in Learning Outcomes by Hours of VR Usage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_of_VR_Usage_Per_Week', y='Improvement_in_Learning_Outcomes', data=data, color='#28A745')
plt.title('Improvement in Learning Outcomes by Hours of VR Usage', fontsize=16)
plt.xlabel('Hours of VR Usage Per Week', fontsize=14)
plt.ylabel('Improvement in Learning Outcomes', fontsize=14)
plt.grid()
plt.show()

# Perceived Effectiveness of VR
plt.figure(figsize=(10, 6))
sns.countplot(x='Perceived_Effectiveness_of_VR', data=data, palette='Set2')
plt.title('Perceived Effectiveness of VR in Education', fontsize=16)
plt.xlabel('Effectiveness Rating', fontsize=14)
plt.ylabel('Number of Responses', fontsize=14)
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:black; overflow:hidden"><b> ‚ú® Stress and Collaboration ‚ú®</b></div>

```python
# Stress Level with VR Usage
plt.figure(figsize=(10, 6))
sns.boxplot(x='Stress_Level_with_VR_Usage', y='Hours_of_VR_Usage_Per_Week', data=data, palette='pastel')
plt.title('Stress Level with VR Usage', fontsize=16)
plt.xlabel('Stress Level', fontsize=14)
plt.ylabel('Hours of VR Usage Per Week', fontsize=14)
plt.show()

# Collaboration with Peers via VR
plt.figure(figsize=(10, 6))
sns.countplot(x='Collaboration_with_Peers_via_VR', data=data, palette='Set2')
plt.title('Impact on Collaboration via VR', fontsize=16)
plt.xlabel('Collaboration Level', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:black; overflow:hidden"><b> ‚ú® Educator Feedback and Instructor Proficiency ‚ú®</b></div>

```python
# Feedback from Educators on VR
plt.figure(figsize=(12, 6))
sns.countplot(x='Feedback_from_Educators_on_VR', data=data, palette='Set2')
plt.title('Feedback from Educators on VR', fontsize=16)
plt.xlabel('Feedback Rating', fontsize=14)
plt.ylabel('Number of Responses', fontsize=14)
plt.xticks(rotation=45)
plt.show()

# Instructor VR Proficiency
plt.figure(figsize=(10, 6))
sns.countplot(x='Instructor_VR_Proficiency', data=data, palette='Set1')
plt.title('Instructor VR Proficiency', fontsize=16)
plt.xlabel('Proficiency Level', fontsize=14)
plt.ylabel('Number of Instructors', fontsize=14)
plt.show()
```

```python
# 7.1 Impact on Creativity by Hours of VR Usage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_of_VR_Usage_Per_Week', y='Impact_on_Creativity', data=data, color='#FF5733')
plt.title('Impact on Creativity by Hours of VR Usage', fontsize=16)
plt.xlabel('Hours of VR Usage Per Week', fontsize=14)
plt.ylabel('Impact on Creativity', fontsize=14)
plt.grid()
plt.show()
```

```python
# 8.1 Importing Required Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import seaborn as sns

# 8.2 Preparing Data for Machine Learning
# Selecting relevant features
features = data[['Hours_of_VR_Usage_Per_Week', 'Engagement_Level', 'Instructor_VR_Proficiency']]
target = data['Improvement_in_Learning_Outcomes']

# Dropping rows with NaN values
features = features.dropna()
target = target[features.index]

# 8.3 One-Hot Encoding for Categorical Features
encoder = OneHotEncoder(sparse=False)
encoded_features = encoder.fit_transform(features)

# 8.4 Label Encoding for Target Variable if it's categorical
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(target)

# 8.5 Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(encoded_features, encoded_target, test_size=0.2, random_state=42)

# 8.6 Training the Model
model = LinearRegression()
model.fit(X_train, y_train)

# 8.7 Making Predictions
y_pred = model.predict(X_test)

# 8.8 Evaluating the Model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')

# 8.9 Visualizing Actual vs Predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='#28A745')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.title('Actual vs Predicted Improvement in Learning Outcomes', fontsize=16)
plt.xlabel('Actual Values', fontsize=14)
plt.ylabel('Predicted Values', fontsize=14)
plt.grid()
plt.show()

# 8.10 Visualizing Residuals
plt.figure(figsize=(10, 6))
sns.residplot(x=y_pred, y=y_test - y_pred, lowess=True, color='blue')
plt.title('Residuals vs Fitted', fontsize=16)
plt.xlabel('Fitted Values', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.axhline(0, color='red', linestyle='--')  # Add a horizontal line at 0 for reference
plt.grid()
plt.show()
```

```python
# 8.12 Evaluating the Model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display metrics
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R¬≤ Score: {r2:.2f}')
```

```python
# 8.13 Visualizing Actual vs Predicted Values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', marker='o', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)  # Diagonal line
plt.title('Actual vs Predicted Values', fontsize=16)
plt.xlabel('Actual Values', fontsize=14)
plt.ylabel('Predicted Values', fontsize=14)
plt.grid()
plt.show()
```
